# ‚úÖ 1. Carregar o CSV no Pandas

```python
import pandas as pd

# Ler o arquivo CSV
df = pd.read_csv("pizza_sales.csv")

# Ver as primeiras linhas
df.head(5)
```

# ‚úÖ 2. Criar banco SQLite e carregar o DataFrame

```python
import sqlite3

# Criar conex√£o com banco em mem√≥ria (ou use um arquivo .db se quiser salvar)
conn = sqlite3.connect(":memory:")

# Enviar o DataFrame para o SQLite
df.to_sql("pizza_sales", conn, index=False, if_exists="replace")
```

```python
# Exemplo: total de vendas por tipo de pizza
query = """
SELECT * 
FROM pizza_sales
LIMIT 5;
"""

df_resultado = pd.read_sql_query(query, conn)
df_resultado
```

# üìä 3. EDA com SQL e Python!

## -> Qual foi o total de receita gerada no per√≠odo analisado?

```python

query = """
SELECT SUM(total_price) AS total_vendas 
FROM pizza_sales;
"""

df_resultado = pd.read_sql_query(query, conn)
df_resultado
```

## -> Qual foi o ticket m√©dio por pedido (valor m√©dio gasto por order_id)?

```python
query = """
SELECT SUM(total_price) / COUNT(DISTINCT order_id) AS ticket_medio
FROM pizza_sales;
"""

df_resultado = pd.read_sql_query(query, conn)
df_resultado
```

## -> Qual foi o ticket m√©dio por pedido (valor m√©dio gasto por order_id)?

```python
query = """
SELECT SUM(total_price) AS total_vendas, SUM(total_price) / COUNT(DISTINCT order_id) AS ticket_medio
FROM pizza_sales;
"""

df_resultado = pd.read_sql_query(query, conn)
df_resultado
```

## -> Quais s√£o as pizzas mais vendidas em quantidade?

```python
query = """
SELECT DISTINCT pizza_name_id, COUNT(order_id) as quantidade_pedidos
FROM pizza_sales
GROUP BY pizza_name_id
ORDER BY quantidade_pedidos DESC;
"""

df_resultado = pd.read_sql_query(query, conn)
df_resultado
```

## -> Quais pizzas geraram mais receita total?

```python

query = """
SELECT DISTINCT pizza_name, SUM(total_price) AS total_vendas 
FROM pizza_sales
GROUP BY pizza_name
ORDER BY total_vendas DESC
LIMIT 10;
"""

df_resultado = pd.read_sql_query(query, conn)
df_resultado
```

```python
query = """
SELECT pizza_name, pizza_size, SUM(total_price) AS receita
FROM pizza_sales
GROUP BY pizza_name
ORDER BY receita DESC
LIMIT 10;
"""

df_resultado = pd.read_sql_query(query, conn)
df_resultado
```

## -> Quais tamanhos de pizza foram mais vendidos?

```python
query = """
SELECT pizza_size, SUM(total_price) AS receita
FROM pizza_sales
GROUP BY pizza_size
ORDER BY receita DESC
LIMIT 10;
"""

df_resultado = pd.read_sql_query(query, conn)
df_resultado
```

## -> Quais tamanhos de pizza foram mais vendidos?

```python
query = """
SELECT pizza_category, SUM(total_price) AS receita
FROM pizza_sales
GROUP BY pizza_category
ORDER BY receita DESC
LIMIT 10;
"""

df_resultado = pd.read_sql_query(query, conn)
df_resultado
```

## -> Em quais hor√°rios do dia ocorrem mais pedidos?

```python

query = """
SELECT order_time, COUNT(order_id) as quantidade_pedidos
FROM pizza_sales
GROUP BY order_time
ORDER BY quantidade_pedidos DESC
LIMIT 10;
"""

df_resultado = pd.read_sql_query(query, conn)
df_resultado
```

## -> Quais s√£o os ingredientes mais comuns nas pizzas mais populares?


```python
query = """
SELECT pizza_ingredients, COUNT(order_id) as quantidade_pedidos
FROM pizza_sales
GROUP BY pizza_ingredients
ORDER BY quantidade_pedidos DESC
LIMIT 10;
"""

df_resultado = pd.read_sql_query(query, conn)
df_resultado
```

## -> Qual √© a m√©dia de pizzas vendidas por pedido?


```python
query = """
SELECT 
SUM(quantity) * 1.0 / COUNT(DISTINCT order_id) AS media_pizzas_por_pedido
FROM pizza_sales;

"""

df_resultado = pd.read_sql_query(query, conn)
df_resultado
```

# ‚úÖ 4. Visualiza√ß√£o Gr√°fica dos Dados!

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Top 5 Pizzas mais vendidas!

# Realizando a query
query = """
SELECT pizza_name, COUNT(order_id) AS quantidade_pedidos
FROM pizza_sales
GROUP BY pizza_name
ORDER BY quantidade_pedidos DESC
LIMIT 5;
"""

top_pizzas = pd.read_sql_query(query, conn)

# Preparando o plot
plt.figure(figsize=(12, 5))  # gr√°fico mais compacto
ax = sns.barplot(
    x=top_pizzas["pizza_name"],
    y=top_pizzas["quantidade_pedidos"],
    hue=top_pizzas["pizza_name"],
    palette="dark:skyblue"
)

# Adicionando os valores em cima das barras
for i, v in enumerate(top_pizzas["quantidade_pedidos"]):
    ax.text(i, v + 20, str(int(v)), ha='center', fontsize=10)

# Personalizando o gr√°fico
plt.title("Top 5 Pizzas Mais Vendidas", fontsize=14)
plt.xlabel("Pizza")
plt.ylabel("Quantidade de Pedidos")
plt.tight_layout()
plt.show()
```

```python
# Top 5 Que Mais Geraram Receita!

query = """
SELECT DISTINCT pizza_name, SUM(total_price) AS total_vendas 
FROM pizza_sales
GROUP BY pizza_name
ORDER BY total_vendas DESC
LIMIT 5;
"""

top_pizzas = pd.read_sql_query(query, conn)

# Preparando o plot
plt.figure(figsize=(14, 5))  # gr√°fico mais compacto
ax = sns.barplot(
    x=top_pizzas["pizza_name"],
    y=top_pizzas["total_vendas"],
    hue=top_pizzas["pizza_name"],
    palette="dark:skyblue",
    legend=False
)

# Adicionando os valores em cima das barras com pontua√ß√£o
for i, v in enumerate(top_pizzas["total_vendas"]):
    valor_formatado = f"{int(v):,}".replace(",", ".")
    ax.text(i, v + 100, valor_formatado, ha='center', fontsize=10)

# Personalizando o gr√°fico
plt.title("Top 5 Pizzas que Mais Geraram Receita", fontsize=14)
plt.xlabel("Pizza")
plt.ylabel("Receita Total (R$)")
plt.tight_layout()
plt.show()
```

```python
# Top Tamanhos de Pizza Mais Vendidos!

# Realizando a query
query = """
SELECT pizza_size, COUNT(order_id) as quantidade_pedidos
FROM pizza_sales
GROUP BY pizza_size
ORDER BY quantidade_pedidos DESC
"""

top_pizzas = pd.read_sql_query(query, conn)

# Preparando o plot
plt.figure(figsize=(12, 5))  # gr√°fico mais compacto
ax = sns.barplot(
    x=top_pizzas["pizza_size"],
    y=top_pizzas["quantidade_pedidos"],
    hue=top_pizzas["pizza_size"],
    palette="dark:skyblue"
)

# Adicionando os valores em cima das barras
for i, v in enumerate(top_pizzas["quantidade_pedidos"]):
    ax.text(i, v + 20, str(int(v)), ha='center', fontsize=10)

# Personalizando o gr√°fico
plt.title("Top Tamanhos de Pizza Mais Vendidos", fontsize=14)
plt.xlabel("Pizza")
plt.ylabel("Quantidade de Pedidos")
plt.tight_layout()
plt.show()
```

```python
# Realizando a query
query = """
SELECT mes, COUNT(order_id) as quantidade_pedidos
FROM pizza_sales
GROUP BY mes
ORDER BY mes
"""

pizzas_por_mes = pd.read_sql_query(query, conn)

# Preparando o gr√°fico de linha
plt.figure(figsize=(10, 5))
ax = sns.lineplot(
    x=pizzas_por_mes["mes"],
    y=pizzas_por_mes["quantidade_pedidos"],
    marker="o",  # adiciona marcadores nos pontos
    color="steelblue"
)

# Adicionando os valores em cima dos pontos
for x, y in zip(pizzas_por_mes["mes"], pizzas_por_mes["quantidade_pedidos"]):
    ax.text(x, y + 10, str(int(y)), ha='center', fontsize=9)

# Personalizando o gr√°fico
plt.title("Pedidos por M√™s", fontsize=14)
plt.xlabel("M√™s")
plt.ylabel("Quantidade de Pedidos")
plt.xticks(range(1, 13))  # garante que os meses apare√ßam de 1 a 12
plt.tight_layout()
plt.show()
```

# üîé Feature Enginneiring! Adicionando novas colunas com base nas ja existestes!

```python
# Checar se a coluna order_date ja esta no formato correto
df.dtypes
```

```python
# Convertendo de Object para Datetime
df["order_date"] = pd.to_datetime(df["order_date"], format="mixed", dayfirst=True, errors="coerce")
```

```python
print(df["order_date"].dtype)
print(df["order_date"].isna().sum())  # quantas datas inv√°lidas foram convertidas em NaT
print(df["order_date"].head())
```

```python
# Criand novas colunas!
df["ano"] = df["order_date"].dt.year
df["mes"] = df["order_date"].dt.month
df["dia"] = df["order_date"].dt.day
df["dia_semana"] = df["order_date"].dt.day_name()  # Segunda, Ter√ßa, etc.
```

```python
# Atualizo as colunas no SQL
df.to_sql("pizza_sales", conn, index=False, if_exists="replace")
```
